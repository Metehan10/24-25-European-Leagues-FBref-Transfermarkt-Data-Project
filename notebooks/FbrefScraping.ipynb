```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# Chrome WebDriver'ı başlat
driver = webdriver.Chrome()
```

```python
# Sayfayı yükle
url = "https://fbref.com/en/comps/10/misc/Championship-Stats"
driver.get(url)
```

```python
# Sayfanın tam yüklenmesi için bekle
time.sleep(5)
```

```python
# Sayfa kaynağını al
html = driver.page_source
```

```python
# BS4 ile parse et
soup = BeautifulSoup(html, 'html.parser')
table = soup.find("table", {"id": "stats_misc"})
```

```python
# DataFrame'e çevir
IngiltereIkinciMiscellaneousStats = pd.read_html(str(table))[0]
driver.quit()
```
